```python
import kagglehub
from matplotlib.pyplot import plot_date

# Download latest version
path = kagglehub.dataset_download("dylanjcastillo/7k-books-with-metadata")

print("Path to dataset files:", path)
```

```python
import pandas as pd
```

```python
books = pd.read_csv(f"{path}/books.csv")
books.head(
    
)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
ax = plt.axes()
sns.heatmap(books.isna().transpose(), cbar=False, ax=ax)

plt.xlabel("Columns")
plt.ylabel("Missing values")
```

```python
import numpy as np

books["missing_description"] = books["description"].isna().astype(int)
books["age_of_book"] = 2025 - books["published_year"]
```

É interesante notar se ha correlação na descricao com as outras variaveis, para nao dropar os missing values e causar vieses para o modelom

```python
columns_of_interest = ["num_pages", "age_of_book", "missing_description", "average_rating"]

correlation_matrix = books[columns_of_interest].corr(method = "spearman")

sns.set_theme(style="white")
plt.figure(figsize=(8, 6))
heatmap = sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", vmin=0, vmax=1, center=0.5)
heatmap.set_title("Correlation Matrix of Columns of Interest")
plt.show()


```

```python
miss_values =books[(books["description"].isna()) |
      (books["num_pages"].isna()) |
      (books["average_rating"].isna()) |
      (books["published_year"].isna())
].shape[0]

miss_values_percent = miss_values / books.shape[0] * 100

print(f"Existem {miss_values} valores ausentes, o que representa {miss_values_percent}% do dataset")
```

303 valores ausentes, o que representa 4% do dataset, nao é um problema para o modelo


```python
book_missing = books[~(books["description"].isna()) &
      ~(books["num_pages"].isna()) &
      ~(books["average_rating"].isna()) &
      ~(books["published_year"].isna())
]
```

```python
book_missing
```

```python
book_missing["categories"].value_counts().reset_index().sort_values(by = "count", ascending = False)
```

```python
plt.figure(figsize=(12, 6))
categories_counts = book_missing["categories"].value_counts()
sns.barplot(x=categories_counts.values[:20], y=categories_counts.index[:20])
plt.title("Distribution of Top 20 Book Categories")
plt.xlabel("Count")
plt.ylabel("Category")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

notase que ha descriçoes com missing values, e que ha descriçoes com poucos caracteres, o que pode ser um problema para o modelo


```python
book_missing["words_in_description"] = book_missing["description"].str.split().str.len()
```

A partir do 25, as descriçoes ja sao longas o suficiente para o modelo

```python
book_missing_25_words = book_missing[book_missing["words_in_description"] >= 25]
book_missing_25_words
```

O subtitle ira ser uma variavel importante para o modelo, pois ha livros com o mesmo titulo e diferentes subititulos


```python
book_missing_25_words["title_and_subtitle"] = (
    np.where(book_missing_25_words["subtitle"].isna(), book_missing_25_words["title"],
             book_missing_25_words[["title", "subtitle"]].astype(str).agg(": ".join, axis = 1))
)

book_missing_25_words["title_and_subtitle"]
```
